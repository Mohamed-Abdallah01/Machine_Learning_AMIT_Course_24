### Drop Un-necessary columns

```python
df.drop(["PassengerId","Name","Ticket"] , axis=1 , inplace=True)
df.head()
```

```python
dtype = df.dtypes
n_unique = df.nunique()

pd.DataFrame({"Dtype":dtype, "Num_unique":n_unique}).T
```

```python
df.columns
```

### change data types

```python
cols = ["Pclass" , "SibSp" ,"Parch","Sex","Embarked","Survived"]
df[cols] = df[cols].astype('category')
pd.DataFrame(df.dtypes).T
```

```python
df["Survived"].dtypes
```

```python
null = df.isnull().sum()
ratio = null / df.shape[0]

pd.DataFrame({"Null_Sum":null , "Ratio":ratio}).T
```

```python
df = df.dropna(subset=["Embarked"])
```

```python
df["Embarked"].isnull().sum()
```

```python
df = df.drop("Cabin",axis=1)
```

```python
plt.figure(figsize=(4,2))
plt.hist(df["Age"] , density=True , edgecolor="black")
plt.title("Age Column Distuributions")
plt.xlabel("Age")
plt.ylabel("Probability")
plt.show()
```

```python
median = df["Age"].median()
df["Age"].fillna(median , inplace=True)
```
