```python
import numpy as np
import matplotlib.pyplot as plt

x=np.array([1,2,3,4])
y=np.array([1,2.8,3.6,4.5])

theta_0=0
theta_1=1
alpha=0.01
num_iterations=1000

sse_values=[]

for i in range(num_iterations):
    y_pred=theta_0+theta_1*x
    d_theta_0=-2*np.sum(y-y_pred)
    d_theta_1=-2*np.sum((y-y_pred)*x)
    
    theta_0 -=alpha*d_theta_0
    theta_1 -=alpha*d_theta_1
    
    sse=np.sum((y-y_pred)**2)
    sse_values.append(sse)
    
    if (i+1)%100==0:
        print(f"iteration {i+1},SSE: {sse}")
    if i==3:print(theta_0,theta_1)    


plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.plot(range(num_iterations), sse_values, label='SSE')
plt.xlabel('Iteration')
plt.ylabel('SSE')
plt.title('SSE over iterations')
plt.legend()

plt.tight_layout()
plt.show()
print(f"optimized parameters: theta_0= {theta_0},theta_1={theta_1}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# البيانات
x = np.array([1, 2, 3, 4])
y = np.array([1, 2.8, 3.6, 4.5])

# إعداد المعاملات
theta_0 = 0  # معامل البداية (الاعتراض)
theta_1 = 0  # معامل البداية (الميل)
alpha = 0.01  # معدل التعلم
num_iterations = 1000  # عدد التكرارات

# قائمة لتخزين SSE لكل تكرار
sse_values = []

# تنفيذ Gradient Descent
for i in range(num_iterations):
    # التنبؤ
    y_pred = theta_0 + theta_1 * x
    
    # حساب التدرج
    d_theta_0 = -2 * np.sum(y - y_pred) / len(x)
    d_theta_1 = -2 * np.sum((y - y_pred) * x) / len(x)
    
    # تحديث المعاملات
    theta_0 -= alpha * d_theta_0
    theta_1 -= alpha * d_theta_1
    
    # حساب SSE
    sse = np.sum((y - y_pred) ** 2)
    sse_values.append(sse)

    # طباعة كل 100 تكرار
    if (i + 1) % 100 == 0:
        print(f"Iteration {i + 1}: SSE = {sse}, theta_0 = {theta_0}, theta_1 = {theta_1}")

# رسم النتائج
plt.plot(range(num_iterations), sse_values)
plt.xlabel('Iteration')
plt.ylabel('SSE')
plt.title('Gradient Descent: SSE vs Iterations')
plt.show()

# عرض المعاملات النهائية
print(f"Final theta_0: {theta_0}, theta_1: {theta_1}")
```
